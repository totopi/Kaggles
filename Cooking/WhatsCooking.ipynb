```python
# Dependencies

# Dataframes and such
import pandas as pd
import numpy as np

# File handling
import json
from pprint import pprint

# Plotting?
import matplotlib.pyplot as plt
```

```python
# Open the training file
file = "train.json"
with open(file) as json_file:
    data = json.load(json_file)
```

```python
# Make a dataframe, ..., ???
cuisine_list = [recipe['cuisine'] for recipe in data]
id_list = [recipe['id'] for recipe in data]
ingredients_list = [recipe['ingredients'] for recipe in data]
food_df = pd.DataFrame({'Cuisine': cuisine_list, 'ID': id_list, 'Ingredients': ingredients_list})
unique_cuisine = food_df['Cuisine'].unique()
cuisine_df = pd.DataFrame({"Cuisine": unique_cuisine})
cuisine_df
```

```python
# idk maybe start with allllll of the ingredients?
ingredient_unique_list = []
food_df["Ingredient Count"] = ""
for index, key in food_df.iterrows():
    for ingr in key["Ingredients"]:
        if ingr not in ingredient_unique_list:
            ingredient_unique_list.append(ingr)
    food_df.set_value(index, "Ingredient Count", len(key["Ingredients"]))
# print(ingredient_unique_list)
```

```python
food_df.head(1)
```

```python
# Maybe it's helpful to have a list of each cuisine's ingredients?
food = []
cuisine_df["Ingredients"] = ""
for index, key in cuisine_df.iterrows():
    for indexs, keys in food_df.loc[food_df["Cuisine"] == key["Cuisine"]].iterrows():
        for ingr in keys["Ingredients"]:
#             if ingr not in food:   uncomment this and indent the line below for ingredients to be unique
            food.append(ingr)
        cuisine_df.set_value(index, "Ingredients", food)
    food = []
```

```python
# As well as the total number of unique ingredients maybe?????

cuisine_df["Ingredient Count"] = ""
for index, key in cuisine_df.iterrows():
    length = len(key["Ingredients"])
    cuisine_df.set_value(index, "Ingredient Count", length)
cuisine_df.head()
```

```python
# Dataframe of cuisine type and count of ingredients: 
data = pd.DataFrame({"Cuisine": unique_cuisine})
for thing in ingredient_unique_list:
    data[thing] = 0
data.head()
```

```python
for index, key in data.iterrows():
    for indexs, keys in food_df.loc[food_df["Cuisine"] == key["Cuisine"]].iterrows():
        for ingr in keys["Ingredients"]:
            count = 0
            count = key[ingr] + 1
            data.set_value(index, ingr, count)
data.head()
```

```python
for index, key in cuisine_df.iterrows():
    for ingr in key["Ingredients"]:
        count = key[ingr] + 1
        cuisine_df.set_value(index, ingr, count)
cuisine_df.head()
```

```python
totalIngList = []
for index, row in cuisine_df.iterrows():
#     cuisineType = row[0]
    ingDict = {}
    ingList = row[1]
    for ingr in ingList:
        if ingr in ingDict:
            ingDict[ingr] = ingDict[ingr] + 1
        else:
            ingDict[ingr] = 1
    totalIngList.append(ingDict)
```
